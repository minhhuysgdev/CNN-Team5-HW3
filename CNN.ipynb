```python
import math

# ==========================
# Bước 1: Khởi tạo tham số
# ==========================

# Input X
X = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Kernel K1
K1 = [
    [1, 0],
    [0, 1]
]

# Bias convolution
b1 = 0

# Fully Connected
W_fc = [0.5, -0.2]

b_fc = [1, 0]

# ==========================
# Bước 2: Convolution
# ==========================

H = len(X)
W = len(X[0])

Kh = len(K1)
Kw = len(K1[0])

output_H = H - Kh + 1
output_W = W - Kw + 1

Z1 = []

for i in range(output_H):

    row = []

    for j in range(output_W):

        s = 0

        for r in range(Kh):
            for c in range(Kw):

                s += X[i+r][j+c] * K1[r][c]

        s += b1

        row.append(s)

    Z1.append(row)

print("Conv Z1")
for row in Z1:
    print(row)

# ==========================
# Bước 3: ReLU
# ==========================

A1 = []

for row in Z1:

    relu_row = []

    for value in row:

        relu_row.append(max(0, value))

    A1.append(relu_row)

print("\nReLU A1")
for row in A1:
    print(row)

# ==========================
# Bước 4: Max Pooling 2x2
# ==========================

max_value = A1[0][0]

for row in A1:
    for value in row:

        if value > max_value:
            max_value = value

P1 = [[max_value]]

print("\nMaxPool P1")
print(P1)

# ==========================
# Bước 5: Flatten
# ==========================

v = []

for row in P1:
    for value in row:
        v.append(value)

print("\nFlatten v")
print(v)

# ==========================
# Bước 6: Fully Connected
# ==========================

s = []

for i in range(len(W_fc)):

    logits = W_fc[i] * v[0] + b_fc[i]

    s.append(logits)

print("\ns (Logits)")
for value in s:
    print(value)

# ==========================
# e^s
# ==========================

exp_s = []

for value in s:

    exp_s.append(math.exp(value))

print("\ne^s")
for value in exp_s:
    print(value)

# ==========================
# Bước 7: Softmax
# ==========================

sum_exp = sum(exp_s)

Y_hat = []

for value in exp_s:

    Y_hat.append(value / sum_exp)

print("\nSoftmax (Y_hat)")
for value in Y_hat:
    print(value)
```
