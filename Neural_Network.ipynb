```python
import numpy as np
import pandas as pd

# ---------------------------------------------------------
# BƯỚC 1: TẢI VÀ CHUẨN BỊ DATASET (UCI Spambase)
# ---------------------------------------------------------
print("Đang tải dataset...")
# Dùng pandas tải trực tiếp dữ liệu từ URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"
df = pd.read_csv(url, header=None)

# Tách Features (X) và Label (Y)
X_raw = df.iloc[:, :-1].values # Lấy tất cả cột trừ cột cuối (57 features đếm từ khóa)
Y = df.iloc[:, -1].values.reshape(-1, 1) # Cột cuối là nhãn (1: Spam, 0: Không Spam)

# Chuẩn hóa dữ liệu (Scale) để Gradient Descent hội tụ nhanh hơn
X_raw = (X_raw - np.mean(X_raw, axis=0)) / np.std(X_raw, axis=0)

N, k = X_raw.shape
print(f"Hoàn tất! Tập dữ liệu có {N} emails, mỗi email có {k} đặc trưng (từ khóa).")
```

```python
# ---------------------------------------------------------
# BƯỚC 2: CÀI ĐẶT NEURAL NETWORK (Mapping với Mã giả)
# ---------------------------------------------------------
epochs = 1000
learning_rate = 0.1

# 1. Khởi tạo W ngẫu nhiên và b = 0
np.random.seed(42)
W = np.random.randn(k, 1) * 0.01
b = 0

print("\nBắt đầu huấn luyện...")
# 2. Vòng lặp huấn luyện
for epoch in range(epochs):
    # --- LAN TRUYỀN XUÔI ---
    Z = np.dot(X_raw, W) + b                  # 3. Z ← X·W + b
    A = 1 / (1 + np.exp(-Z))                  # 4. Hàm kích hoạt Sigmoid

    # --- LAN TRUYỀN NGƯỢC ---
    dZ = A - Y                                # 5. Tính sai số
    dW = (1/N) * np.dot(X_raw.T, dZ)          # 6. Đạo hàm trọng số
    db = (1/N) * np.sum(dZ)                   # 7. Đạo hàm bias

    # --- CẬP NHẬT TRỌNG SỐ ---
    W -= learning_rate * dW                   # 8. Cập nhật W
    b -= learning_rate * db                   # 9. Cập nhật b

    if epoch % 200 == 0:
        # Tính Loss để theo dõi (công thức Binary Cross-Entropy)
        # Cộng thêm eps (1e-15) để tránh lỗi log(0)
        eps = 1e-15
        loss = -np.mean(Y * np.log(A + eps) + (1 - Y) * np.log(1 - A + eps))
        print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")
```

```python
# ---------------------------------------------------------
# BƯỚC 3: ĐÁNH GIÁ MÔ HÌNH
# ---------------------------------------------------------
print("\n--- KẾT QUẢ HUẤN LUYỆN ---")
# Email nào có xác suất A > 0.5 thì dự đoán là Spam (1)
Y_pred = (A > 0.5).astype(int)

# Tính độ chính xác (Accuracy)
accuracy = np.mean(Y_pred == Y) * 100
print(f"Độ chính xác trên toàn bộ tập dữ liệu: {accuracy:.2f}%")
```
